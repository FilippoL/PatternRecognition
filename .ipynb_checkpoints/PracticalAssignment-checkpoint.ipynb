# MNIST Digit Recognition: Exploratory Data Analysis 

## Report Description

During this report the Modified National Institute of Standards and Technology data set is going to be explored. 

```r
# Load CSV file in
mnist.dat <- read.csv("mnist.csv")

# Import libraries
library(OpenImageR) 
library(dplyr)
library(iterators)
```

```r
# Show image for testing
imageShow(matrix(as.numeric(mnist.dat[1,-1]),nrow=28,ncol=28,byrow=T))
```

```r
# Print data set shape
dim(mnist.dat[,-1])

# Plot count of each digit's pixel in barplot and then display it in a table
barplot(table(mnist.dat[,1]), col=rainbow(10, 0.5), main="Digits in dataset")
as.data.frame(mnist.dat) %>% group_by(label) %>% summarise(count = n())
```
