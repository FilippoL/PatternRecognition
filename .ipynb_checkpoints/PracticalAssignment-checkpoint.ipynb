```r
# Load CSV file in
mnist.dat <- read.csv("mnist.csv")
```

```r
# Show image for testing
imageShow(matrix(as.numeric(mnist.dat[1,-1]),nrow=28,ncol=28,byrow=T))
```

```r
mnist.summary = data.frame(t(sapply( mnist.dat[,-1] , function(x) cbind(mean = mean(x) ,
                                  sd = sd(x) ,
                                  median = median(x) ,
                                  minimum = min(x) ,
                                  maximum = max(x) ,
                                  s.size = length(x)))))
colnames(mnist.summary) = c("mean","sd","median","minimum","maximum","size")
```

```r
mnist.summary[mnist.summary$sd==0,]
```

### These pixels have no information.  With standard deviation of 0.

```r
indices_of_useless_features = which(mnist.summary$sd == 0)
actual_digit = mnist.dat[90,-1]
actual_digit[,indices_of_useless_features] = 125
digit_example = matrix(as.numeric(actual_digit),nrow=28,ncol=28,byrow=T)
imageShow(digit_example)
```

### These pixels have little information. Below standard deviation of 1. (NOISE)

```r
indices_of_useless_features = which(mnist.summary$sd < 1)
actual_digit = mnist.dat[400,-1]
actual_digit[,indices_of_useless_features] = 125
digit_example = matrix(as.numeric(actual_digit),nrow=28,ncol=28,byrow=T)
imageShow(digit_example)
```

```r
# Plot count of each digit's pixel in barplot and then display it in a table
label_distribution = table(mnist.dat[,1]) 
barplot(label_distribution, col=rainbow(10, 0.5), main="Digits in dataset")

expected_accuracy_on_majority_label_classification = label_distribution[2] / sum(label_distribution)
sprintf("Accuracy if 1 was predicted for all labels: %.2f %%", expected_accuracy_on_majority_label_classification * 100)
# as.data.frame(mnist.dat) %>% group_by(label) %>% summarise(count = n())
```

```r
clean <- function(digit_image){
    actual_digit = digit_image
    indices_of_useful_features = which(actual_digit < 248)
    actual_digit[,indices_of_useful_features] = 0
    digit_example = matrix(as.numeric(actual_digit),nrow=28,ncol=28,byrow=T)
    imageShow(digit_example)
}


cleaned_data = mnist.dat
apply(cleaned_data[,-1], MARGIN=1, FUN=clean)
```

```r
density_per_image = apply(mnist.dat[,-1], MARGIN=1, FUN=function(x) sum(x)/784)
plot(mnist.dat$label, density_per_image, main="Density per digit", xlab="Digit", ylab="Ink USed")
```

```r
summary_per_digit = rbind.data.frame(tapply(density_per_image, mnist.dat[,1],  function(x) cbind(mean = mean(x) ,
                                  sd = sd(x) ,
                                  median = median(x) ,
                                  minimum = min(x) ,
                                  maximum = max(x),
                                  s.size = length(x))
                           ))
rownames(summary_per_digit) = c("mean","sd","median","minimum","maximum","size") 
#TODO: Plot all distributions
summary_per_digit
```

```r
pixels_data = mnist.dat[,-1]
scaled_pixels_data = scale(pixels_data)
scaled_pixels_data[is.na(scaled_pixels_data)] = 0 #Replacing NaN with 0
```

```r
class(density_per_image)
length(density_per_image)
```

```r
scaled_density = scale(density_per_image)
scaled_density[is.na(scaled_density)] = 0 #Replacing NaN with 0
data = cbind.data.frame(row.names = NULL, density=scaled_density, label=mnist.dat$label)

head(data)

plot(density(boxplot.stats(data$density)$out), main="Outliers density")

plot(data$label, data$density, main="Pixel values per Digit", xlab="Digit", ylab="Digit pixel values")

hist(as.numeric(data$density), col=rainbow(10, 0.5), main="Frequency of scaled pixel values", xlab="Number of pixels")
```

```r
density_model = multinom(data$label ~ ., data, maxit=5000)
```

```r
density_model
```

```r
predicted = predict(density_model, data, type="class")
table(data$label, predicted)
plot(density(as.numeric(predicted)))
```
